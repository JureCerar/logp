```python
import numpy as np
import matplotlib.pyplot as plt

def fb(x, x0=0., r=0., k=1000.):
    """Flat-bottom potential"""
    dxr = abs(x - x0) - abs(r)
    if k > 0:
        return 0.5 * k * dxr**2 * np.heaviside(dxr, 1)
    else:        
        return -0.5 * k * dxr**2 * np.heaviside(-dxr, 1)

# Constants
r = 1.0
k = 1000

x = np.linspace(-2.0, 2.0, 301)
y1 = fb(x, 0, r,  k)
y2 = fb(x, 0, r, -k)

plt.plot(x, y1, label="Normal")
plt.plot(x, y2, label="Inverted")
plt.legend() 
```
